```python
data = pd.read_csv("/kaggle/input/housing-data/Parcels_2015.csv", thousands=',')
data.head(3)
```

```python
data.hist(bins=50, figsize=(10,10))
```

```python
data.drop("TAX_MAP", axis=1, inplace=True)
data.drop("OWNER_NAME", axis=1, inplace=True)
data.drop("Shape_STLe", axis=1, inplace=True)
data.drop("Shape_ST_1", axis=1, inplace=True)
data.drop("Shape_ST_2", axis=1, inplace=True)
data.drop("Shape_STAr", axis=1, inplace=True)
data.drop("ZIP_POSTAL", axis=1, inplace=True)
data.drop("the_geom", axis=1, inplace=True)
data.drop("PIN", axis=1, inplace=True)
data.drop("MAP", axis=1, inplace=True)
data.drop("BLOCK", axis=1, inplace=True)
data.drop("LOT", axis=1, inplace=True)
data.drop("ADDRESS1", axis=1, inplace=True)
data.drop("ADDRESS2", axis=1, inplace=True)
data.drop("FREE_LINE_", axis=1, inplace=True)
data.drop("P_ID", axis=1, inplace=True)
data.drop("YEAR_ID", axis=1, inplace=True)
data.drop("PROPERTY_A", axis=1, inplace=True)

data.head(3)

data_train=data
```

```python
data_train['LAND_VALUE'].replace(0, data['LAND_VALUE'].median(), inplace=True)
data_train['TOTAL_VALU'].replace(0, data['TOTAL_VALU'].median(), inplace=True)
data_train['ACRES'].replace(0, data['ACRES'].median(), inplace=True)
data_train['BUILDING_V'].replace(0, data['BUILDING_V'].median(), inplace=True)
data_train.head(3)
```
